```python
#Question 1
def count(s):
    v = 0
    c = 0
    for i in s:
        if i in ['a','e','i','o','u']:
            v = v+1
        else:
            c = c+1
    return v,c
       
s = input("Enter a string ")
v,c = (count(s))
print("Number of vowels : ",v)
print('Number of consonants : ',c)


#Question2
def mult():
   
    r1 = int(input('Enter number of rows for m1: '))
    c1 = int(input('Enter number of columns for m1: '))
    r2 = int(input('Enter number of rows for m2: '))
    c2 = int(input('Enter number of columns for m2: '))
    
    if c1 != r2:
        print("Error: Number of columns of m1 must be equal to number of rows of m2")
        return

    print("Enter elements for matrix m1:")
    m1 = []
    for i in range(r1):
        a = []
        for j in range(c1):
            v = int(input())
            a.append(v)
        m1.append(a)
    
    print("Enter elements for matrix m2:")
    m2 = []
    for i in range(r2):
        row = []
        for j in range(c2):
            v = int(input())
            row.append(v)
        m2.append(row)
    
    result = [[0 for _ in range(c2)] for _ in range(r1)]
    
    for i in range(r1):
        for j in range(c2):
            sum = 0
            for k in range(c1):
                sum += m1[i][k] * m2[k][j]
            result[i][j] = sum
    
    print("Result of matrix multiplication:")
    for row in result:
        print(row)

mult()


#Question 3
def common():
    l1 = []
    l2 = []
    s1 = int(input("Enter size of list 1"))
    s2 = int(input('Enter size of list 2'))

    for i in range(s1):
        v = int(input())
        l1.append(v)

    for i in range(s2):
        v = int(input())
        l2.append(v)
    
    l3 = []
    for i in l1:
        if i in l2:
            l3.append(i)
            
    return l3

print('The common elements are : ',common())


#Question 4
def transpose():
    r = int(input('Enter number of rows'))
    c = int(input('Enter number of columns'))
    m = []
    for i in range(r):
        a = []
        for j in range(c):
            v = int(input())
            a.append(v)
        m.append(a)
    print(m)

    t = [[m[i][j] for i in range(r)]for j in range(c)]
    for i in range(r):
        for j in range(c):
            if i != j:
                t[i][j]=m[j][i]
    return t

print(transpose())
```
